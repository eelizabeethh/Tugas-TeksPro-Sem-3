```python
import requests
from bs4 import BeautifulSoup

root = 'https://www.rottentomatoes.com/browse/movies_in_theaters/sort:popular?page=3'

title = []
dates = []

url = requests.get(root)
soup = BeautifulSoup(url.text, 'html.parser')

for t in soup.find_all('span', {'data-qa':'discovery-media-list-item-title'}):
    list_title = t.text.strip()
    title.append(list_title)

for d in soup.find_all('span', {'data-qa':'discovery-media-list-item-start-date'}):
    list_dates = d.text.strip()
    dates.append(list_dates)

for i in range(len(dates)):
    print(f"Title: {title[i]}, Realease Date: {dates[i]}")
```

```python
len(title), len(dates)
```

```python
import pandas as pd
min_length = min(len(title), len(dates))

title = title[:min_length]
dates = dates[:min_length]

df = pd.DataFrame(
    {'Title': title,
     'Realease Date': dates,
     })

df.head(84)
```
